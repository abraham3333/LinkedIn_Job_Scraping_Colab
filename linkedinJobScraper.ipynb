```python
#Import Packages
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
```

```python

# Chrome setting
chrome_options = Options()
chrome_options.add_argument("--headless")  # headless mod
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--remote-debugging-port=9222")  # Debugging port

# ChromeDriver service
driver = webdriver.Chrome(options=chrome_options)

# LinkedIn job page
driver.get('https://www.linkedin.com/jobs/search/?currentJobId=3982665032&distance=25&f_WT=2%2C1%2C3&geoId=102424322&keywords=data%20scientist&origin=JOBS_HOME_KEYWORD_HISTORY&refresh=true'
)

# Wait for the page to load
time.sleep(5)

# Find job postings
job_listings = driver.find_elements(By.XPATH, '//*[@id="main-content"]/section[2]/ul/li')

# Print job postings
for listing in job_listings:
    try:
        job_title_element = listing.find_element(By.CLASS_NAME, 'sr-only')
        job_title = job_title_element.text
        print("Job Title:", job_title)

        address_element = listing.find_element(By.CLASS_NAME, 'base-search-card__metadata')
        address = address_element.text
        print("Address:", address)
        link_element = listing.find_element(By.CLASS_NAME, 'base-card__full-link')
        link = link_element.get_attribute('href')
        print("Job Link:", link)
        print ("-" * 50)
    except Exception as e:
        print("Error:", e)

# Close browser
driver.quit()
```
